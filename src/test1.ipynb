```python
markets_coordinates.dtypes
```

```python
markets_coordinates['x'] = [x.replace(',', '.') for x in markets_coordinates['x']]
markets_coordinates['y'] = markets_coordinates['y'].map(lambda x: x.replace(',', '.'))
```

```python
markets_coordinates
```

```python
markets_coordinates = markets_coordinates.astype({
  'id': 'str',
  'x': 'float32',
  'y': 'float32'
})
markets_coordinates.dtypes
```

```python
markets_coordinates.describe()
```

```python
markets_coordinates.shape
```

```python
markets_coordinates.head(5)
```

```python
markets_coordinates_gdf = gpd.GeoDataFrame(
  data=markets_coordinates[['id']],
  geometry=gpd.points_from_xy(markets_coordinates['x'],markets_coordinates['y']),
  crs={'init':'epsg:4326'}
)
markets_coordinates_gdf
```
